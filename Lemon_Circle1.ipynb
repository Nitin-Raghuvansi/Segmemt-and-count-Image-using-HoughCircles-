```
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('lemons1.jpeg')
#img = cv2.medianBlur(img,5)
img1 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img1)
plt.show()
#img = cv2.medianBlur(img,5)
circles = cv2.HoughCircles(img1,cv2.HOUGH_GRADIENT,1, 42, np.array([]), 80, 20, 3, 40)

circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(img,(i[0],i[1]),i[2],(0,255,0),2)
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,str(circles.shape[1])+' lemons detected',(10,50), font, 0.8,(255,255,255),2,cv2.LINE_AA)
print(str(circles.shape[1]))
plt.imshow(img)  
```
